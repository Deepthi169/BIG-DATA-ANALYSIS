```python
import pandas as pd
import plotly.express as px
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(list(uploaded.keys())[0])
```

```python
print("First 5 rows of dataset:")
print(df.head())
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
fig1 = px.scatter(df, x='age', y='claim', color='smoker',
                  size='bmi', hover_data=['region', 'diabetic'],
                  title="Age vs Claim Amount (Colored by Smoker Status, Sized by BMI)")
fig1.show()
```

```python
avg_claim_region = df.groupby('region', as_index=False)['claim'].mean()
fig2 = px.bar(avg_claim_region, x='region', y='claim', color='region',
              title="Average Claim Amount by Region")
fig2.show()
```

```python
fig3 = px.box(df, x='smoker', y='claim', color='smoker',
              title="Claim Distribution for Smokers vs Non-Smokers")
fig3.show()
```

```python
avg_bp_claim = df.groupby('bloodpressure', as_index=False)['claim'].mean()
fig4 = px.line(avg_bp_claim, x='bloodpressure', y='claim',
               title="Average Claim vs Blood Pressure")
fig4.show()
```

```python
avg_claim_gender_diabetic = df.groupby(['gender', 'diabetic'], as_index=False)['claim'].mean()
fig5 = px.bar(avg_claim_gender_diabetic, x='gender', y='claim', color='diabetic', barmode='group',
              title="Average Claim by Gender and Diabetic Status")
fig5.show()
```
